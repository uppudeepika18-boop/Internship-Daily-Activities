# Freature creation

```python
import pandas as pd
df=pd.DataFrame({
    "study hours":[2,4,6,8,10],
    "attendance":[60,70,80,90,95],
    "maths":[40,55,65,75,80],
    "science":[42,58,68,78,88]
    
})
df
```

# domain-based features

```python
df["total-marks"] =df["maths"] + df["science"]
df
```

# Mathematical Feature

```python
df["marks_per_hour"] =df["total-marks"]/df["study hours"]
df
```

# Interaction Feature

```python
df["study_attendance_interaction"] =df["study hours"] + df["attendance"]
df
```

# Polynomial Feature

```python
df["study_hour_squared"] = df["study hours"]
df
```

```python
corr =df.corr()["science"].abs() ## abs covenrts -ve values into +ve values
corr
```
