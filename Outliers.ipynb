# IQR Method

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
LB = Q1 - 1.5 * IQR
UB = Q3 + 1.5 * IQR
outliers = data[(data<(LB))|(data>(UB))]
```

# Z_Score

```python
import pandas as pd
data=pd.DataFrame({
    "hour":[1,2,3,4,5,6,7,8,20]
})
data
```

```python
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
```

```python
zscores=np.abs(stats.zscore(data["hour"]))
out=data[zscores>2]
out
```

```python
stats.zscore(data["hour"])
```

```python
zscor=np.abs(stats.zscore(data["hour"]))
zscor
```

```python
plt.scatter(range(len(data)),data["hour"])
plt.xlabel("index")
plt.ylabel("hours")
plt.title("scatter plot")
plt.show()
```

```python
plt.hist(data["hour"],bins=9)
plt.xlabel("study")
plt.ylabel("frncy")
plt.title("histogram plot")
plt.show()
```
