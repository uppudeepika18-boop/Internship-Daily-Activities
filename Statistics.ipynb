# concate Data Frame

```python
india_weather = pd.DataFrame({
    "city" : ["mumbai","hyd","delhi"],
    "Temp":[32,33,35]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city" : ["nyc","chicago","orland"],
    "Temp":[32,23,25]
})
us_weather
```

```python
df =pd.concat([india_weather,us_weather])
df
```

```python
df =pd.concat([india_weather,us_weather],ignore_index=True)  #ignore index is used to change index in order
df
```

```python
df =pd.concat([india_weather,us_weather],axis=1)
df
```

# Merge Data Frames

```python
temp = pd.DataFrame({
    "city" : ["mumbai","hyd","delhi","vizag"],
    "Temp":[32,33,35,40]
})
temp
```

```python
humidity = pd.DataFrame({
    "city" : ["bengaluru","hyd","chennai","tirupati"],
    "humidity":[2,33,42,40]
})
humidity
```

```python
df =pd.merge(temp,humidity,on="city")
df
```

```python
df =pd.merge(temp,humidity,on="city",how="outer")
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df['Temp']=df['Temp'].fillna(df.Temp.min())
df.isnull()
```
