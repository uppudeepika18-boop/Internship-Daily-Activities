# Binning

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    "age":[18,22,25,30,35,40,45,50]
})
print(data)
```

# Equal-Width

```python
data["age_bin"] = pd.cut(data["age"],bins=4)
data
```

```python
plt.hist(data["age"], bins=4)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Equal-Width Binning using pd.cut()")
plt.show()
```

```python
plt.scatter(range(len(data)),data["age"])
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Equal-Width Binning using pd.cut()")
plt.show()
```

```python
data["age_frequency_bin"] = pd.qcut(data["age"], q=4)
data
```
