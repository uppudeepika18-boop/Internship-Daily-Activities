```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame({
    "study_hours":[1,2,3,4,5],
    "marks":[2,4,5,4,5]
})
df
```

```python
X=df[["study_hours"]]
Y=df[["marks"]]
```

```python
model=LinearRegression()
model.fit(X,Y)
```

```python
X_line = pd.DataFrame(
    {'study_hours': np.linspace(X.min()[0], X.max()[0], 100)}
)
Y_line = model.predict(X_line)
```

```python
X_line
```

```python
print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)
```

```python
X_new=pd.DataFrame({'study_hours':[4.0]})
Y_new=model.predict(X_new)
print("Predict Value: ",Y_new)
```

```python
plt.figure()
plt.scatter(X,Y)
plt.plot(X_line, Y_line)
plt.scatter(X_new,Y_new)
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Linear Regression: Best-Fit Line")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

Y_pred = model.predict(X)
mse = mean_squared_error(Y, Y_pred)

print("MSE:", mse)
```

# Multiple Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df=pd.read_csv(r"D:\MLR_Study_marks.csv")
df
```

```python
X = df[["Study_Hours", "Attendance", "Sleep_Hours"]]
Y = df["Marks"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state = 40)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)

print("\nModel Intercept:", model.intercept_)
print("Model Coefficients:", model.coef_)
```

```python
Y_pred = model.predict(X_test)
```

```python
plt.scatter(Y_test, Y_pred)
plt.xlabel("Actual Marks")
plt.ylabel("Predicted Marks")
plt.title("Actual vs Predicted Marks")
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'r--')
plt.show()
```

```python
Y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
print("\nMean Sqaured Error:", mse)
print("R^2 Score:", r2)
```
